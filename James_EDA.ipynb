```python
import pandas as pd
df = pd.read_csv("train.csv")
```

```python
# 📊 Initial Exploration
print("Shape:", df.shape)
print(df.info())
print(df.describe())
print("Missing values:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())
print("Unique values:\n", df.nunique())
```

```python

#  Univariate Analysis

# Age distribution
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()
# Embarked count
sns.countplot(x='Embarked', data=df)
plt.title("Passengers by Embarked Port")
plt.show()

# Fare distribution
sns.boxplot(x=df['Fare'])
plt.title("Fare Distribution (with outliers)")
plt.show()

```

```python

#  Bivariate Analysis

# Fare vs Pclass
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare by Passenger Class")
plt.show()

# Age vs Survival
sns.histplot(data=df, x="Age", hue="Survived", kde=True)
plt.title("Survival by Age")
plt.show()

# Embarked vs Survival
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival by Embarked Location")
plt.show()
```

```python

#  Outlier Detection & Handling
# Visualize outliers in Fare
sns.boxplot(x=df['Fare'])
plt.title("Fare Outliers")
plt.show()

# Option: remove extreme fares
q_hi = df['Fare'].quantile(0.99)
df_filtered = df[df['Fare'] < q_hi]
```

```python

#  Multivariate Analysis
# Pclass, Age, and Fare vs Survived
sns.catplot(x='Pclass', y='Age', hue='Survived', col='Embarked', data=df)
plt.suptitle("Multivariate Survival Analysis")
plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------------
#  Target Variable Analysis - Survived
# -----------------------------------------------
sns.countplot(x='Survived', data=df)
plt.title("Survival Counts")
plt.show()

# Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()

# Survival by Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```
